# Download sea ice concentration
This script downloads the sea ice concentration:

from https://data.seaice.uni-bremen.de/amsr2/asi_daygrid_swath/n6250/netcdf/2026/

Use this to dwownload and extract online NetCDF data files.

```python
# Modules
import subprocess
from datetime import datetime, timedelta
from pathlib import Path

newdate = datetime.now()
newdate.strftime('%Y%m%d')
```

```python
delta = timedelta(days = 1)
delta
```

```python
# Configuration
BASE_URL = ("https://data.seaice.uni-bremen.de/amsr2/asi_daygrid_swath/n6250/netcdf/2026/")
OUTPUT_DIR = Path("amsr2_feb_2026_nc")
OUTPUT_DIR.mkdir(parents = True, exist_ok = True)
START_DATE = datetime(2026, 2, 1)
END_DATE = datetime(2026, 2, 28)
```

```python
newdate = START_DATE + delta
newdate
```

```python
# Download loop
current_date = START_DATE
while current_date <= END_DATE:
    date_str = current_date.strftime("%Y%m%d")
    filename = f"asi-AMSR2-s6250-{date_str}-v5.4.nc"
    url = BASE_URL + filename
    output_file = OUTPUT_DIR / filename
    cmd = ["curl","-o", str(output_file),url]
    result = subprocess.run(cmd)
    current_date += timedelta(days=1)
```
